```python
import pandas as pd
import numpy as np
import yfinance as yf
from plotly.subplots import make_subplots
import plotly.graph_objects as go

get_ipython().run_line_magic("matplotlib", "inline")
import warnings
warnings.filterwarnings("ignore")
```

```python
# Carregando a base de dados e fazendo os checks iniciais

# Carrega as bases
ticker1 = "^BVSP"
df1 = yf.download(ticker1, "2012-01-01", "2022-12-31")
df1["Retornos"] = df1["Adj Close"].pct_change(1)
```

```python
df1.head()
```

```python
# Como temos datas no índice é simples fazer a filtragem

df1_12 = df1["2012"]
df1_13 = df1["2013"]
df1_14 = df1["2014"]
df1_15 = df1["2015"]
df1_16 = df1["2016"]
df1_17 = df1["2017"]
df1_18 = df1["2018"]
df1_19 = df1["2019"]
df1_20 = df1["2020"]
df1_21 = df1["2021"]
df1_22 = df1["2022"]
```

```python
df1_12["Normalizado"] = 100*(df1_12 / df1_12.iloc[0,:])["Adj Close"]
df1_13["Normalizado"] = 100*(df1_13 / df1_13.iloc[0,:])["Adj Close"]
df1_14["Normalizado"] = 100*(df1_14 / df1_14.iloc[0,:])["Adj Close"]
df1_15["Normalizado"] = 100*(df1_15 / df1_15.iloc[0,:])["Adj Close"]
df1_16["Normalizado"] = 100*(df1_16 / df1_16.iloc[0,:])["Adj Close"]
df1_17["Normalizado"] = 100*(df1_17 / df1_17.iloc[0,:])["Adj Close"]
df1_18["Normalizado"] = 100*(df1_18 / df1_18.iloc[0,:])["Adj Close"]
df1_19["Normalizado"] = 100*(df1_19 / df1_19.iloc[0,:])["Adj Close"]
df1_20["Normalizado"] = 100*(df1_20 / df1_20.iloc[0,:])["Adj Close"]
df1_21["Normalizado"] = 100*(df1_21 / df1_21.iloc[0,:])["Adj Close"]
df1_22["Normalizado"] = 100*(df1_22 / df1_20.iloc[0,:])["Adj Close"]
```

```python
df1_12.head()
```

```python
fig = make_subplots(rows = 1, cols = 1,
                    shared_xaxes = True,
                    vertical_spacing = 0.08)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_12["Normalizado"],
                         name = "2012", line = dict(color = "blue")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_13["Normalizado"],
                         name = "2013", line = dict(color = "red")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_14["Normalizado"],
                         name = "2014", line = dict(color = "orange")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_15["Normalizado"],
                         name = "2015", line = dict(color = "purple")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_16["Normalizado"],
                         name = "2016", line = dict(color = "black")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_17["Normalizado"],
                         name = "2017", line = dict(color = "green")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_18["Normalizado"],
                         name = "2018", line = dict(color = "fuchsia")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_19["Normalizado"],
                         name = "2019", line = dict(color = "gray")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_20["Normalizado"],
                         name = "2020", line = dict(color = "dark blue")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_21["Normalizado"],
                         name = "2021", line = dict(color = "dark green")),
                         row = 1, col = 1)

fig.add_trace(go.Scatter(x = df1_12.index, y = df1_22["Normalizado"],
                         name = "2022", line = dict(color = "dark gray")),
                         row = 1, col = 1)




fig.update_layout(height = 600, width = 600,
                  title_text = "Performance Anual " + ticker1,
                  font_color = "blue",
                  title_font_color = "black",
                  xaxis_title = "Meses",
                  xaxis = dict(tickformat = "%m"),
                  yaxis_title = "Fechamento Normalizado",
                  legend_title = "Anos",
                  font = dict(size = 15, color = "Black")
                  )
fig.update_layout(hovermode="x")

fig.show()
```
