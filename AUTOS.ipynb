```python
def clean_col(col):
    col = col.replace("yearOfRegistration","registration_year")
    col = col.replace("monthOfRegistration","registration_month")
    col = col.replace("notRepairedDamage","unrepaired_damage")
    col = col.replace("yearOfRegistration","registration_year")  
    col = col.replace("dateCrawled","date_crawled")
    col = col.replace("lastSeen","last_seen")
    col = col.replace("dateCreated","ad_created")
    return col

new_columns = []
for c in autos.columns:
    clean_c = clean_col(c)
    new_columns.append(clean_c)

ac = new_columns
autos.columns = ac
```

```python
autos["price"] = autos["price"].str.replace('$','')
autos["price"] = autos["price"].str.replace(',','')
autos["odometer"] = autos["odometer"].str.replace('km','')
autos["odometer"] = autos["odometer"].str.replace(',','')
autos["price"] = autos["price"].astype(int)
autos["odometer"] = autos["odometer"].astype(int)
```

```python
best_brands = autos["brand"].value_counts().head(6).index #o problema pediu 6
best_brands
```

```python
bmp = {} 

for b in best_brands: 
    mean_price = autos[autos["brand"] == b]["price"].mean() 
    bmp[b] = int(mean_price)

bmp
```

```python
bmps = pd.Series(bmp)
print(bmps)
```

```python
bmpDF = pd.DataFrame(bmps, columns=['mean_price'])
bmpDF = bmpDF.sort_values("mean_price", ascending=False)
bmpDF
```

```python
mileage = {} 

for b in best_brands: 
    mean_price = autos[autos["brand"] == b]["odometer_km"].mean() 
    mileage[b] = int(mean_price)

mileage
mileages = pd.Series(mileage)
bmpDF['mileage'] = mileages
bmpDF
```

```python
bmpDF = bmpDF.sort_values("mileage", ascending=False)
bmpDF
```
